```python
#pandas ques. 001
# Employee DataFrame operations (a to i)

import pandas as pd

# Creating the initial DataFrame
data = {
    "Name": ["Atul", "Raj", "Darpan", "Anmol", "Piyush"],
    "Age": [25, 30, 23, 22, 32],
    "Designation": ["Manager", "Clerk", "Analyst", "Clerk", "Manager"],
    "Salary": [56000, 25000, 35000, 28000, 58000],
    "Bonus": [15000, 7000, 9000, 13000, 12000]
}

Employee = pd.DataFrame(data, index=["E101", "E102", "E103", "E104", "E105"])
print("Original DataFrame:\n", Employee)

# a) Add a new row (FIXED: 5 values for 5 columns)
Employee.loc["E106"] = ["GEET", 28, "Analyst", 70000, 15000]
print("\nAfter adding new row:\n", Employee)

# b) Add a column called 'Shift' with 'Morning'
Employee["Shift"] = "Morning"
print("\nAfter adding Shift column:\n", Employee)

# c) Print details of youngest employee
youngest = Employee[Employee["Age"] == Employee["Age"].min()]
print("\nYoungest Employee:\n", youngest)

# d) Print details of all clerks
clerks = Employee[Employee["Designation"] == "Clerk"]
print("\nAll Clerks:\n", clerks)

# e) Print average Bonus
avg_bonus = Employee["Bonus"].mean()
print("\nAverage Bonus:", avg_bonus)

# f) Print total salary distributed
total_salary = Employee["Salary"].sum()
print("\nTotal Salary Distributed:", total_salary)

# g) Add column 'City'
Employee["City"] = ["Janak Puri", "Tilak Nagar", "Shiv Nagar",
                     "Vikas Puri", "Janak Puri", "Tilak Nagar"]
print("\nAfter adding City column:\n", Employee)

# h) Delete records of E103 and E105
Employee = Employee.drop(["E103", "E105"])
print("\nAfter deleting E103 and E105:\n", Employee)

# i) Rename column 'Bonus' to 'Incentive'
Employee = Employee.rename(columns={"Bonus": "Incentive"})
print("\nAfter renaming Bonus to Incentive:\n", Employee)
```

```python
#pandas ques 002
# PART 1: DataFrame operations

import pandas as pd

# Creating the DataFrame
data = {
    "Count": [15, 20, 25, 35],
    "Price": [250, 200, 600, 900]
}
data_df = pd.DataFrame(data, index=["Pencil", "Pen", "Eraser", "Scale"])
print("Original DataFrame:\n", data_df)

# a) Change Eraser count to 25 (already 25, but done again)
data_df.loc["Eraser", "Count"] = 25

# b) Rename index 'Scale' as 'Wooden Scale'
data_df = data_df.rename(index={"Scale": "Wooden Scale"})

# c) Change row labels
data_df.index = ["Product 1", "Product 2", "Product 3", "Product 4"]

# d) Change column names
data_df.columns = ["Discount", "Total"]

# e) Add new row 'Product 5'
data_df.loc["Product 5"] = [10, 100]

print("\nAfter all modifications:\n", data_df)

# f) Create another dataframe 'Products' without column 'Total'
Products = data_df.drop(columns=["Total"])
print("\nProducts DataFrame (without Total):\n", Products)
```

```python
#pandas ques 002
# PART 2: Agriculture DataFrame

import pandas as pd
import numpy as np

data2 = {
    "Fruits": [44, 11950, 7152, 140169, 7830, 113],
    "Pulses": [23, 818, 33, 3184, 931, 1],
    "Rice": [814, 1930, 11586, 13574, 7452, 2604],
    "Wheat": [5, 2737, 16440, 30056, np.nan, np.nan]
}

states = ["Tripura", "Gujrat", "Punjab", "UP", "AP", "Kerala"]
agri = pd.DataFrame(data2, index=states)
print("\nAgriculture DataFrame:\n", agri)

# a) Total production for each state
print("\nTotal production for each state:")
print(agri.sum(axis=1))

# b) Total production for each item type
print("\nTotal production for each item:")
print(agri.sum())

# c) How many states produce Wheat?
print("\nNumber of states producing Wheat:")
print(agri["Wheat"].count())
```
